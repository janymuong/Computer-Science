By examining carbon dioxide levels, we can gauge the extent of greenhouse gas emission, a primary driver of global warming and we can gain a holistic understanding of the state of our planet's climate and its impacts on both natural and human systems.

## `.describe()` - Describe Tool and Statistics

The `.describe()` method in pandas generates descriptive statistics of a **DataFrame**. It summarizes statistics for numerical columns, such as *count*, *mean*, *standard deviation*, *minimum*, *maximum*, etc.

```python
df_low_carbon.describe()
```

```python
# include all stats for both numerical and categorical columns.
# the include='all' parameter ensures that statistics for all types of columns are part of the output

df_low_carbon.describe(include='all')
```

```python
# get stats similar as pandas .descibe() tool but LESS ans using a for loop;
for col in df_low_carbon.columns:
    if is_numeric_dtype(df_low_carbon[col]):
        print(f'{col}:')
        print(f'\t Mean = {df_low_carbon[col].mean():.2f}')
        print(f'\t Standard Deviation = {df_low_carbon[col].std():.2f}')
        print(f'\t Minimum Value = {df_low_carbon[col].min():.2f}')
        print(f'\t Maximum Value = {df_low_carbon[col].max():.2f}')
```

### Filter the DataFrame - Columns For Low Carbon Adoption

These columns provide data on the trade, investment, emissions, and environmental aspects related to low carbon technologies and practices. Analyzing these columns can help evaluate the level of adoption and the effectiveness of low carbon initiatives in each country or region.

```python
# filter the DataFrame to include only relevant columns related to low carbon adoption;
relevant_columns = ['Country', 'ISO3', 'Indicator', 'Unit', 'CTS_Full_Descriptor', 'F2013', 'F2014', 'F2015', 'F2016', 'F2017', 'F2018', 'F2019', 'F2020', 'F2021', 'F2022']
df_low_carbon_rel = df_low_carbon[relevant_columns]

# verify the relevant columns
print(df_low_carbon_rel.head())

# assess low carbon adoption based on specific indicators
# example 1: imports of low carbon technology products as percent of total trade
imports_indicator = df_low_carbon_rel[df_low_carbon_rel['Indicator'].str.contains('Imports of low carbon technology products')]
imports_values = imports_indicator[['Country', 'ISO3', 'Indicator', 'F2013', 'F2014', 'F2015', 'F2016', 'F2017', 'F2018', 'F2019', 'F2020', 'F2021', 'F2022']]
print('Imports of low carbon technology products:')
print(imports_values)

# example 2: trade balance in low carbon technology products in US dollars
trade_balance_indicator = df_low_carbon_rel[df_low_carbon_rel['Indicator'].str.contains('Trade balance in low carbon technology products')]
trade_balance_values = trade_balance_indicator[['Country', 'ISO3', 'Indicator', 'F2013', 'F2014', 'F2015', 'F2016', 'F2017', 'F2018', 'F2019', 'F2020', 'F2021', 'F2022']]
print('Trade balance in low carbon technology products:')
print(trade_balance_values)
```

## Pandas Visualizations
We can create high-quality figures and graphics for the Low Carbon Dataset and also	plot and make sexy visualizations from the results.
